```python
import numpy as np
import matplotlib.pyplot as plt

# Constantes:
m_e = 9.10938356 * 10**(-31)
hbar1 = 1.0545718 * 10**(-14)
hbar2 = 6.58211915 * 10**(-16)
hbar = hbar1*hbar2
cte = np.power(hbar,2)/(2*m_e)
delta_L = 8 ; L1 = L2 = 5; V1 = V2 = 4
Ei = 0.0001 ; Ef = 12

E = list() ; R = list() ; T = list() ; RT = list()

for i in np.arange(Ei+0.01,Ef + 0.01,0.01):

  if V1 != V2 and L1 != L2:
    k1 = k3 = k5 = np.sqrt((i)/cte)
    if i > V1 and  i > V2:
      k2 = np.sqrt(((i-V1)/cte))
      k4 = np.sqrt(((i-V2)/cte))
    elif i > V1 and i < V2:
      k2 = np.sqrt(((i-V1)/cte))
      k4 = np.sqrt(((V2-i)/cte))*1j
    elif i < V1 and i > V2:
      k2 = np.sqrt(((V1-i)/cte))*1j
      k4 = np.sqrt(((i-V2)/cte))
    elif i < V1 and i < V2:
      k2 = np.sqrt(((V1-i)/cte))*1j
      k4 = np.sqrt(((V2-i)/cte))*1j

  elif V1 != V2 and L1 == L2:
    k1 = k3 = k5 = np.sqrt((i/cte))
    if i > V1 and i > V2:
      k2 = np.sqrt(((i-V1)/cte))
      k4 = np.sqrt(((i-V2)/cte))
    elif i > V1 and i < V2:
      k2 = np.sqrt(((i-V1)/cte))
      k4 = np.sqrt(((V2-i)/cte))*1j
    elif i < V1 and i > V2:
      k2 = np.sqrt(((V1-i)/cte))*1j
      k4 = np.sqrt(((i-V2)/cte))
    elif i < V1 and i < V2:
      k2 = np.sqrt(((V1-i)/cte))*1j
      k4 = np.sqrt(((V2-i)/cte))*1j

  elif V1 == V2 and L1 != L2:
    k1 = k3 = k5 = np.sqrt((i/cte))
    if i > V1 and i > V2:
      k2 = np.sqrt(((i-V1)/cte))
      k4 = np.sqrt(((i-V2)/cte))
    elif i > V1 and i < V2:
      k2 = np.sqrt(((i-V1)/cte))
      k4 = np.sqrt(((V2-i)/cte))*1j
    elif i < V1 and i > V2:
      k2 = np.sqrt(((V1-i)/cte))*1j
      k4 = np.sqrt(((i-V1)/cte))
    elif i < V1 and i < V2:
      k2 = np.sqrt(((V1-i)/cte))*1j
      k4 = np.sqrt(((V2-i)/cte))*1j

  elif V1 == V2 and L1 == L2:
    k1 = k3 = k5 = np.sqrt((i/cte))
    if i > V1 and i > V2:
      k2 = np.sqrt(((i-V1)/cte))
      k4 = np.sqrt(((i-V2)/cte))
    elif i > V1 and i < V2:
      k2 = np.sqrt(((i-V1)/cte))
      k4 = np.sqrt(((V2-i)/cte))*1j
    elif i < V1 and i > V2:
      k2 = np.sqrt(((V1-i)/cte))*1j
      k4 = np.sqrt(((i-V2)/cte))
    elif i < V1 and i < V2:
      k2 = np.sqrt(((V1-i)/cte))*1j
      k4 = np.sqrt(((V2-i)/cte))*1j

  # Matriz A(8,8).
  A = np.zeros((8,8),dtype = np.cdouble())

  # Primeira linha:
  A[0,0] = 1 ; A[0,1] = -1 ; A[0,2] = -1
  # Segunda linha:
  A[1,1] = np.exp(1j*k2*L1) ; A[1,2] = np.exp(-1j*k2*L1) ; A[1,3] = -np.exp(1j*k3*L1)  ; A[1,4] = -np.exp(-1j*k3*L1)
  # Terceira linha:
  A[2,3] = np.exp(1j*k3*(L1 + delta_L)) ; A[2,4] = np.exp(-1j*k3*(L1 + delta_L)) ; A[2,5] = -np.exp(1j*k4*(L1 + delta_L)) ; A[2,6] = -np.exp(-1j*k4*(L1 + delta_L))
  # Quarta linha:
  A[3,5] = np.exp(1j*k4*(L1 +delta_L+ L2)) ; A[3,6] = np.exp(-1j*k4*(L1 +delta_L+ L2))   ; A[3,7] = -np.exp(1j*k1*(L1 +delta_L+ L2))
  # Quinta linha:
  A[4,0] = -k1 ; A[4,1] = -k2 ; A[4,2] = k2
  # Sexta linha:
  A[5,1] = k2*np.exp(1j*k2*L1) ; A[5,2] = -k2*np.exp(-1j*k2*L1) ; A[5,3] = -k3*np.exp(1j*k3*L1)  ; A[5,4] = k3*np.exp(-1j*k3*L1)
  # Sétima linha:
  A[6,3] = k3*np.exp(1j*k3*(L1 + delta_L)) ; A[6,4] = -k3*np.exp(-1j*k3*(L1 + delta_L)) ; A[6,5] = -k4*np.exp(1j*k4*(L1 + delta_L)) ; A[6,6] = k4*np.exp(-1j*k4*(L1 + delta_L))
  # # Oitava linha:
  A[7,5] = k4*np.exp(1j*k4*(L1 +delta_L+ L2)) ; A[7,6] = -k4*np.exp(-1j*k4*(L1 +delta_L+ L2))  ; A[7,7] = -k1*np.exp(1j*k1*(L1 +delta_L+ L2))

  # Matriz b(8,1):
  b = np.zeros((8,1),dtype = np.cdouble())
  b[0,0] = -1 ; b[4,0] = -k1

  # Solução do sistema linear:
  x = np.linalg.solve(A,b)

  # Coeficiente de Reflexão:
  r = x[0,0]*np.conj(x[0,0])
  R.append(r.real)

  # Coeficiente de Transmissão:
  t = x[7,0]*np.conj(x[7,0])
  T.append(t.real)

  # Soma dos coeficientes:
  RT.append(R+T)

  # Energia:
  E.append(i)


print('R =',R)
print('T =',T)
print('RT =',RT)

plt.figure()
plt.plot(E,T,color='Red',linestyle='-.',label='T')
plt.plot(E,R,color='Blue',linestyle='-.',label='R')
plt.plot(E,RT,color='Black',linestyle='-',label='R(E)+T(E)')
plt.xlabel('E')
plt.ylabel('R(E) e T(E)')
plt.xlim([0,12])
plt.ylim([0,1.1])
plt.title('Potencial Barreira Dupla')
plt.legend(loc='best')
plt.grid(True)
plt.show()
```
